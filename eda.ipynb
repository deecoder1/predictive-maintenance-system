```python
# Assuming 'data' is your DataFrame

# Function to count outliers in a column using the IQR method
def count_outliers(column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers)

# List of pressure columns
pressure_cols = ['Lub oil pressure', 'Fuel pressure', 'Coolant pressure']

# Count and print outliers for each column
for col in pressure_cols:
    num_outliers = count_outliers(col)
    print(f"Number of outliers in {col}: {num_outliers}")
```

Lub oil pressure: 66 outliers (0.34% of 19,535 rows).
Fuel pressure: 1135 outliers (5.8% of rows).
Coolant pressure: 785 outliers (4% of rows).
